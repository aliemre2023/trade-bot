```python
all_news = ""
for folder in os.listdir("news"):
    for file in os.listdir(f"news/{folder}"):
        f = open(f"news/{folder}/{file}")
        all_news += f.read()
```

```python
for unwanted in unwanted_contents:
    all_news = all_news.replace(unwanted,"")
```

```python
word_dict = dict()
for word in all_news.split(" "):
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1
```

```python
word_dict = dict(sorted(word_dict.items(), key=lambda item: item[1], reverse=True))
word_dict = {k: v for k, v in word_dict.items() if 80 < v < 90}
print(len(word_dict))
```

```python
plt.bar(word_dict.keys(),word_dict.values())
plt.xticks(rotation=90)
plt.show()
```
